```python
import os
import shutil
import zipfile
```

```python
# Define the paths
dogs_vs_cats_zip_path = 'data/zip file/dogs-vs-cats.zip'
output_folder_path_zip = 'data/zip file'

# Extract the dogs-vs-cats.zip file
with zipfile.ZipFile(dogs_vs_cats_zip_path, 'r') as zip_ref:
    zip_ref.extractall(output_folder_path_zip)
    
print('dogs-vs-cats.zip File Extraction Completed')

# Define the path to the train.zip file
output_folder_path = 'data/processed data'
train_zip_path = 'data/raw data/train.zip'

# Extract the train.zip file
with zipfile.ZipFile(train_zip_path, 'r') as zip_ref:
    zip_ref.extractall(output_folder_path)

print('train.zip File Extraction Completed')

# Define the path to the train.zip file
test_zip_path = 'data/raw data/test1.zip'

# Extract the train.zip file
with zipfile.ZipFile(test_zip_path, 'r') as zip_ref:
    zip_ref.extractall(output_folder_path)

print('test1.zip File Extraction Completed')

print("\nAll File Extraction Completed")
```

```python
# Path to the "train" folder in Google Drive
train_folder = 'data/processed data/train'

# Destination folders for cat and dog images
cat_folder = 'data/processed data/train/cat'
dog_folder = 'data/processed data/train/dog'

# Create the cat and dog folders if they don't exist
os.makedirs(cat_folder, exist_ok=True)
os.makedirs(dog_folder, exist_ok=True)

print('Cat and Dog Folder Created')

# Iterate over the files in the train folder
cat = 0
dog = 0
unknown = 0

for filename in os.listdir(train_folder):
    # Construct the source and destination paths
    src_path = os.path.join(train_folder, filename)

    # Skip directories
    if os.path.isdir(src_path):
        continue

    # Extract the class label from the filename
    label = filename.split('.')[0]

    # Define the destination folder based on the class label
    if label == 'cat':
        dest_folder = cat_folder
        cat+=1
    elif label == 'dog':
        dest_folder = dog_folder
        dog+=1
    else:
        unknown += 1
        continue  # Skip files with unrecognized labels

    # Move the image to the respective cat or dog folder
    shutil.move(src_path, dest_folder)

print("Images moved to the respective cat and dog folders.\n")

print(f"Total Images: {cat+dog+unknown}\n\t=> Total Dog Images: {dog}\n\t=> Total Cat Images: {cat}\n\t=> Total Unknown Images: {unknown}")
```
